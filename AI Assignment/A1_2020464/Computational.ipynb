```python
Destinations = [
    {
        "name": "Jaipur",
        "weather": "Hot",
        "budget": 28000,
        "activities": ["Visit Amber Fort", "Explore City Palace", "Shopping in Johari Bazaar"],
        "rating": 7,
        "feedback": [],
    },
    {
        "name": "Shimla",
        "weather": "Cold",
        "budget": 27000,
        "activities": ["Mall Road Shopping", "Jakhu Temple", "Kufri Skiing"],
        "rating": 8,
        "feedback": [],
    },
    {
        "name": "Mumbai",
        "weather": "Warm",
        "budget": 33000,
        "activities": ["Gateway of India", "Marine Drive", "Bollywood Studio Tour"],
        "rating": 8,
        "feedback": [],
    },
    {
        "name": "Kerala",
        "weather": "Tropical",
        "budget": 40000,
        "activities": ["Backwaters Cruise", "Ayurveda Treatment", "Visit to Hill Stations"],
        "rating": 9.5,
        "feedback": [],
    },
    {
        "name": "Rajasthan",
        "weather": "Hot",
        "budget": 45000,
        "activities": ["Visit to Forts and Palaces", "Camel Safari", "Desert Camping"],
        "rating": 9,
        "feedback": [],
    },
    {
        "name": "Tamil Nadu",
        "weather": "Tropical",
        "budget": 35000,
        "activities": ["Visit to Temples", "Beaches", "Hill Stations"],
        "rating": 8.5,
        "feedback": [],
    },
    {
        "name": "Karnataka",
        "weather": "Tropical",
        "budget": 40000,
        "activities": ["Visit to Temples", "National Parks", "Hill Stations"],
        "rating": 9,
        "feedback": [],
    },
]


# Here, we will create a function to find the matching_destinations
# The time complexity of this function is O(n) because it iterates through the numbers of  Destinations list, which has n elements, where has n number of destinations.
def find_matching_destinations(weather_preferred, budget):
    destination_matching = []

    for destination in Destinations:
        if destination['weather'].lower() == weather_preferred.lower() and destination["budget"] <= budget:
            destination_matching.append(destination["name"])

    return destination_matching


# We Created a new function to recommend the user to suggest the destinations
#Time complexity is O(n) because it depends on find_matching_destinations which has time complexity O(n) 
def recommend_destination():
    weatherPreferred = input("Enter the Preferred weather that you want to visit to ")
    budget = int(input("Enter your budget for the destination"))

    matching_destinations = find_matching_destinations(weatherPreferred, budget)

    if not matching_destinations:
        print("No matching destinations found.")
    else:
        print("Recommended places:", ", ".join(matching_destinations))


# We create a new function in which users can add more destinations
# Time complexcity here is O(1) as it just adds new destinations to the dictionary
def add_new_destination():
    name_of_destination = input("Enter the name of the destination in which you want to add ")
    ratings = input("Enter the rating out of 10 ")

    Destinations.append({
        "name": name_of_destination,
        "rating": ratings,
        "feedback": []
    })

    print(Destinations)


# Here, we create a function to add feedback
# Time complexity here is O(n) because it iterates through the Destinations list, which has n number of elements. 
# If the destination is not found, time complexity is O(1) 
# Overall, the worst time complexity is O(n)
def add_feedback():
    nameDes = input("Write  the name of the destination you want to give feedback on to ")
    commentDest = input("Write the feedback ")

    # To ensure if name of destination exits in the Destination Dictionary or not
    destination_found = False

    for destination in Destinations:
        if (destination["name"].lower() == nameDes.lower()):
            destination["feedback"].append({"comment": commentDest})
            destination_found = True
            print("Added successfully.")
            break

    if not destination_found:
        print("Destination not found.")

    print(Destinations)


while 1:
    print("\nMenu:")
    print("1. Recommend Destination")
    print("2. Add New Destination")
    print("3. Add Feedback")
    print("4. Exit")

    enterTheoption = int(input("Enter the option 1,2,3,4 "))

    if enterTheoption == 1:
        recommend_destination()
    elif enterTheoption == 2:
        add_new_destination()
    elif enterTheoption == 3:
        add_feedback()
    elif enterTheoption == 4:
        break
    else:
        print("Invalid option")

print("Thanks")


#Overall, this code has the worst time complexity of O(n)
```
