# Cross Validation 

```python
# Initializijng the variable to store R2 scores
R2Scores = []

# Number of sets to run
num_of_iterations = 20

# Initializing the index variable
i = 0


while i < num_of_iterations:
    # Here for training test_size is 0.3 because in the question train test split is 70 percent
    # Means 70% goes for training and rest 30% for testing
    XTrain, XTemp, yTrain, yTemp = train_test_split(X, y, test_size=0.3, random_state=None)
    # Here test_size is 0.3 because train value and test split are given 15% combines then to 30 %
    # Now the remaing will be 70-30 = 40% now for better result in which data to be split equally 
    # We will take 50% means test_size is 0.5
    XVal, XTest, yVal, yTest = train_test_split(XTemp, yTemp, test_size=0.5, random_state=None)

    # Create a linear regression model
    model = LinearRegression()

    # Train the model on the training set
    model.fit(XTrain, yTrain)

    # Make predictions using the validation set
    yValPred = model.predict(XVal)

    # Calculating R2 score for the validation set
    valR2 = r2_score(yVal, yValPred)

    yTestPred = model.predict(XTest)

    # Calculating R2 score for the testing set
    testR2 = r2_score(yTest, yTestPred)

    # Append the R2 score to the list
    R2Scores.append(testR2)

    # Increment the index
    i += 1

# Calculating the average and standard deviation of R2 scores
average_r2 = np.mean(R2Scores)
std_dev_r2 = np.std(R2Scores)

print(f'Average R2 Score over {num_sets} sets: {average_r2:.2f}')
print(f'Standard Deviation of R2 Score: {std_dev_r2:.2f}')
```
